```python
# Importing libraries
import requests
```

```python
# Function to send a request to the weather API and print weather information
def request(URL):
    # Sending a GET request to the provided URL
    req = requests.get(url=URL)
    # Parsing the JSON response
    req_result = req.json()
    
    # Extracting weather condition, description, and temperature from the response
    condition = req_result['weather'][0]['main']
    description = req_result['weather'][0]['description']
    temp = req_result['main']['temp']
    
    # Printing the weather information
    print(f"Condition: {condition} \nTemp: {temp}'c \nDescription: {description}\n")

# Function to get weather by city name
def weather_city(city, appid):
    # Constructing the URL with the city name and API key
    URL = f'https://api.openweathermap.org/data/2.5/weather?q={city}&units=metric&appid={appid}'
    # Making a weather request using the constructed URL
    request(URL)

# Function to get weather by zip code
def weather_zip(zipcode, appid):
    # Constructing the URL with the zip code and API key
    URL = f'https://api.openweathermap.org/data/2.5/weather?zip={zipcode}&units=metric&appid={appid}'
    # Making a weather request using the constructed URL
    request(URL)

# Function to get weather by latitude and longitude
def weather_lat_long(lat, lon, appid):
    # Constructing the URL with latitude, longitude, and API key
    URL = f"https://api.openweathermap.org/data/2.5/weather?lat={lat}&lon={lon}&appid={appid}"
    # Making a weather request using the constructed URL
    request(URL)
    
# Opening and reading the API key from a file
with open('appid.txt', 'r') as file:
    appid = file.readline().strip()

# Welcome message
print("Welcome to Chimdi and Akunna's Weather App")
print("Select any of the options to Check Weather in your desired location: \n 1. city \n 2. zip code \n 3. Longitude and Latitude \n 4. Close app")

# Main function to interact with the user and call appropriate weather functions
def main():
    while True:
        try:
            # Taking user input for options
            user_input = int(input("Enter your an option between 1-4: "))
            # Checking user choice and calling respective functions
            if user_input == 1:
                city = input("Enter city: ")
                weather_city(city, appid)
            elif user_input == 2:
                zipcode = input("Enter zipcode: ")
                weather_zip(zipcode, appid)
            elif user_input == 3:
                lat = input("Enter latitude: ")
                lon = input("Enter longitude: ")
                weather_lat_long(lat, lon, appid)
            elif user_input == 4:
                break
            else:
                print("Enter a valid input")
        except ValueError:
            print("Please enter a valid integer between 1-4.")

# Calling the main function to start the program
main()

# Asking the user if they want to check the weather again
print("\nDo you want to check the weather again? \n Y for Yes \n Any other character for No")
response = input("Enter option: ")
if response.lower() == "y":
    main()

# Thank you message
print("\nThank you for using the Weather App.")
```
